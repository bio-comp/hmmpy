# Introduction to Hidden Markov Models

This notebook introduces the fundamental concepts of Hidden Markov Models (HMMs) based on the Rabiner (1989) tutorial.

## What is a Hidden Markov Model?

An HMM is a statistical model where the system is assumed to be a Markov process with **hidden states**. We can only observe the outputs (observations), not the hidden states.

### The Markov Chain

A **Markov chain** is a stochastic process where the probability of being in a future state depends only on the current state, not on how we arrived there:

$$P(q_t = S_j | q_{t-1} = S_i, q_{t-2} = S_k, ...) = P(q_t = S_j | q_{t-1} = S_i)$$

This is called the **Markov property** or the **first-order assumption**.

### Key Components

| Symbol | Meaning | Shape |
|--------|---------|-------|
| $N$ | Number of hidden states | scalar |
| $M$ | Number of observation symbols | scalar |
| $\mathbf{A}$ | State transition probability matrix | $N \times N$ |
| $\mathbf{B}$ | Observation emission probability matrix | $N \times M$ |
| $\oldsymbol{\pi}$ | Initial state distribution | $N$ |

### The Three Fundamental Questions

1. **Evaluation**: Given an observation sequence $O$ and model $\lambda$, what is $P(O|\lambda)$?
   - Solved by: **Forward algorithm**

2. **Decoding**: Given an observation sequence $O$, what is the most likely state sequence $Q$?
   - Solved by: **Viterbi algorithm**

3. **Learning**: How do we adjust model parameters $\lambda = (\mathbf{A}, \mathbf{B}, \boldsymbol{\pi})$ to maximize $P(O|\lambda)$?
   - Solved by: **Baum-Welch algorithm** (EM)

## Creating an HMM

Let's create a simple HMM with 2 hidden states and 6 observation symbols.

```python
# Define the model parameters
A = np.array([[0.95, 0.05], [0.05, 0.95]])  # Transition matrix
B = np.array([[1/6]*6, [1/10]*5 + [1/2]])    # Emission matrix
V = [1, 2, 3, 4, 5, 6]                      # Observation symbols

# Create the HMM
hmm = HMM(n_states=2, A=A, B=B, V=V)

print("Number of states:", hmm.N)
print("Number of symbols:", hmm.M)
print("\nTransition Matrix A:")
print(hmm.A)
print("\nEmission Matrix B:")
print(hmm.B)
print("\nInitial State Distribution:")
print(hmm.Pi)
```

## Random Initialization

If you don't provide matrices, HMM will randomly initialize them.

```python
# Create HMM with random initialization
hmm_random = HMM(n_states=3, V=[0, 1, 2])

print("Random Transition Matrix (rows sum to 1):")
print(hmm_random.A)
print("\nSum of each row:", hmm_random.A.sum(axis=1))
```

## Fixed Emission Probabilities

You can fix certain states to always emit specific symbols.

```python
# Create HMM with fixed emissions
A = np.array([[0.7, 0.3], [0.3, 0.7]])
F = {0: np.array([1.0, 0.0, 0.0])}  # State 0 always emits symbol 0

hmm_fixed = HMM(n_states=2, A=A, V=[0, 1, 2], F=F)

print("Emission matrix with fixed state 0:")
print(hmm_fixed.B)
print("\nFixed emissions:", hmm_fixed.F)
```

## Visualizing HMM Parameters

```python
import matplotlib.pyplot as plt

# Create a simple HMM for visualization
A = np.array([[0.7, 0.3], [0.3, 0.7]])
B = np.array([[0.5, 0.3, 0.2], [0.2, 0.3, 0.5]])
hmm_viz = HMM(n_states=2, A=A, B=B, V=[0, 1, 2])

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# Transition matrix heatmap
im0 = axes[0].imshow(hmm_viz.A, cmap='Blues', vmin=0, vmax=1)
axes[0].set_xticks([0, 1])
axes[0].set_yticks([0, 1])
axes[0].set_xticklabels(['State 0', 'State 1'])
axes[0].set_yticklabels(['State 0', 'State 1'])
axes[0].set_title('Transition Matrix A\n(P(next state | current state))')
axes[0].set_xlabel('To State')
axes[0].set_ylabel('From State')
for i in range(2):
    for j in range(2):
        axes[0].text(j, i, f'{hmm_viz.A[i, j]:.2f}', ha='center', va='center', fontsize=12)
plt.colorbar(im0, ax=axes[0])

# Emission matrix heatmap
im1 = axes[1].imshow(hmm_viz.B, cmap='Greens', vmin=0, vmax=1)
axes[1].set_xticks([0, 1, 2])
axes[1].set_yticks([0, 1])
axes[1].set_xticklabels(['Sym 0', 'Sym 1', 'Sym 2'])
axes[1].set_yticklabels(['State 0', 'State 1'])
axes[1].set_title('Emission Matrix B\n(P(observation | state))')
axes[1].set_xlabel('Observation')
axes[1].set_ylabel('State')
for i in range(2):
    for j in range(3):
        axes[1].text(j, i, f'{hmm_viz.B[i, j]:.2f}', ha='center', va='center', fontsize=12)
plt.colorbar(im1, ax=axes[1])

# Initial distribution
axes[2].bar(['State 0', 'State 1'], hmm_viz.Pi, color=['steelblue', 'coral'])
axes[2].set_title('Initial State Distribution Pi')
axes[2].set_ylabel('Probability')
axes[2].set_ylim(0, 1)

plt.tight_layout()
plt.show()

print("Each row in A and B sums to 1 (valid probability distributions)")
```
