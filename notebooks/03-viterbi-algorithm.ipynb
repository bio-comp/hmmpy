# Viterbi Algorithm

The Viterbi algorithm finds the most likely sequence of hidden states that produced an observation sequence. This is the **decoding problem** in HMMs.

## Mathematical Background

The Viterbi algorithm is similar to the forward algorithm but uses **max** instead of **sum** to find the most likely path.

### Key Variables

- $\delta_t(i)$: Maximum probability of being in state $i$ at time $t$ with the most likely path
- $\psi_t(i)$: Backpointer to the previous state that led to $\delta_t(i)$

### Recursion (Rabiner Eq. 37-38)

**Initialization** (t=1):
$$
\delta_1(i) = \pi_i b_i(o_1), \quad \psi_1(i) = 0
$$

**Induction** (t = 2, 3, ...):
$$
\delta_t(j) = \max_{1 \leq i \leq N} [\delta_{t-1}(i) a_{ij}] b_j(o_t)
$$

$$
\psi_t(j) = \arg\max_{1 \leq i \leq N} [\delta_{t-1}(i) a_{ij}]
$$

**Termination**:
$$
P^* = \max_{1 \leq i \leq N} \delta_T(i), \quad q_T^* = \arg\max_{i} \delta_T(i)
$$

```python
import numpy as np

from hmm import HMM, forward, viterbi
```

## Basic Viterbi Decoding

```python
# Create an HMM with biased transitions
A = np.array([[0.9, 0.1], [0.1, 0.9]])  # Strong self-transitions
B = np.array([[0.7, 0.1, 0.2], [0.2, 0.7, 0.1]])  # Different emission patterns
V = [0, 1, 2]

hmm = HMM(n_states=2, A=A, B=B, V=V)

# Observation sequence: mostly 0s with some 1s
obs = [0, 0, 0, 1, 0, 0, 2, 0, 0]

# Run Viterbi
states, delta, psi = viterbi(hmm, obs, scaling=True)

print(f"Observation: {obs}")
print(f"Most likely states: {states}")
print("\nDelta matrix (max probabilities):")
print(np.exp(delta))  # Convert from log scale
```

## Comparison with Forward

Viterbi finds the most likely path, while Forward computes the total probability.

```python
# Compare Viterbi with Forward
log_prob_forward, _, _ = forward(hmm, obs, scaling=True)
log_prob_viterbi = delta[states[-1], len(obs)-1]

print(f"Forward log probability: {log_prob_forward:.4f}")
print(f"Viterbi path log probability: {log_prob_viterbi:.4f}")
print("\nNote: Viterbi probability <= Forward probability")
print("(Forward sums over all paths, Viterbi uses only the best path)")
```

## Visualizing State Probabilities

```python
import matplotlib.pyplot as plt

# Get forward probabilities for comparison
log_prob, alpha, c = forward(hmm, obs, scaling=True)
state_probs = alpha  # Approximate state probabilities

fig, ax = plt.subplots(figsize=(10, 4))
t = range(len(obs))
ax.plot(t, state_probs[0, :], 'b-o', label='State 0')
ax.plot(t, state_probs[1, :], 'r-o', label='State 1')
ax.plot(t, [0.5]*len(obs), 'k--', alpha=0.3, label='Threshold')
ax.set_xlabel('Time Step')
ax.set_ylabel('State Probability')
ax.set_title('State Probabilities Over Time')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()

print("\nViterbi path states:", states)
```

## Summary

- Viterbi finds the most likely state sequence given observations
- Uses dynamic programming (delta and psi matrices)
- Viterbi probability <= Forward probability (sum vs max)
